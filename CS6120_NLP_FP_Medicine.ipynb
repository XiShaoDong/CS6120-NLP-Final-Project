```python
import pandas as pd

df = pd.read_csv("Medicine_Details.csv")

print("Columns in dataset:", df.columns.tolist())

composition_unique = df['Composition'].nunique()
print(f"Number of unique Composition: {composition_unique}")

print("Sample Composition:", df['Composition'].unique()[:10])

uses_unique = df['Uses'].nunique()
print(f"Number of unique Uses: {uses_unique}")

print("Sample Uses:", df['Uses'].unique()[:10])
```
